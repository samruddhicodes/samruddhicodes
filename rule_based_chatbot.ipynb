```python
import re

print("Chatbot service started. Type 'bye' to exit.")

while True:
    user = input("You:").strip().lower()
    if re.search(r"\b(bye|exit|quit)\b",user):
        print("Bot: Goodbye! keep learning and improving yourself!")
        break
    elif re.search(r"\b(hi|hello|hey)\b",user):
        print("Bot: Hi! I am Sam , your virtual assistant .How can i help you today ?")
    elif re.search(r"\b(about yourself|your name)\b",user):
        print("Bot:I am Sam,here to guide you on self-help books,learning tips,tech projects,internships.")
    elif re.search(r"\b(book|self-help|read|recommend)\b",user):
        print("Bot: That sounds interesting ! What types of books do you usually enjoy reading?")
    elif re.search(r"\b(tech|skills|project|internship)\b",user):
        print("Bot: Practice small projects ,explore tech blogs , keep your skills updated daily.")
    elif re.search(r"\b(help|guide|support)\b",user):
        print("Bot: Sure!You can ask me about self-help books,tech tips as well as for learning strategies.")
    else:
        print("Bot: I am afraid I didn't quite grasp that. To make sure I help you correctly , could you provide a bit more detail?")
```
